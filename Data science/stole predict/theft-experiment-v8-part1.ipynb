## Section 1: Introduction

iLSP (Inteligencia Logística en Seguridad Privada) is a Mexican security logistics company that provides a security management service for their customers that are shipping goods across the country via land using trucks. Customers of iLSP include companies such as Mars, Procter&Gamble, Unilever, DHL, and Mining companies. Given the large size of the country, journeys can last up to 2 days or longer, and may travel through dangerous and extremely remote zones.


<img src="./img/trucks.jpg" width=500/>


In Mexico, is it not uncommon for these vehicles carrying large amounts of goods to be stolen, either partially or complete theft of the trailer, and sometimes involving violence. The stolen goods are usually resold on the black/grey markets, and as such can be any products, for example shampoos, iPhones, or clothes. Stolen goods are usually consumer goods, rather than raw materials, given that they are easy to sell without raising suspicion.

iLSP claims that a theft that is detected within 8 minutes is far more likely to be prevented or recovered, compared to a theft that is not detected for 8 minutes or more. In fact, in the SLEA (Service Level Execution Agreement) that iLSP signs with customers, thefts must be detected within a certain time frame (depending on the customer) to avoid financial repercussions for iLSP. As such, iLSP’s main business objectives are:

1. **Prevention**: reduce the likelihood of a theft event by understanding security risk factors and ensuring that all trips take best practices into account
2. **Detection**: detect theft events as rapidly as possible (within minutes)
3. **Reaction**: take appropriate actions in order to recover stolen goods and ensure safety of Operators.

To achieve this, iLSP provides a service that involves various components:

* **Monitoreo:** Real-time tracking via a team of Monitoristas, who continuously monitor journeys in real-time to detect any anomalies, for example deviation from the pre-designated route or unusually long stops, and keep in regular contact with the Operators (truck drivers) to understand why. Any potential thefts should be detected within minutes, allowing the Monitoring team to handover the journey to the Crisis team.
* **Mesa de Crisis:** Crisis-management team that coordinates iLSP security patrols in the field, as well as local and federal Police, to prevent or recover stolen goods in real-time. If a theft has been detected within 8 minutes of starting, the probability of preventing the theft or recovering stolen goods is greatly increased.
* **Custodia:** iLSP employee(s) that physically accompany the trucks on their journey, in order to provide more reliable and up-to-date information that is closely integrated with iLSP. A Custodia may be following the truck in a car or motorbike, or sitting in the truck with the driver.
* **Patrullas:** iLSP security patrols in the field that patrol known dangerous zones, in order to provide rapid response to potential thefts. Additional charge.
* **Consulting:** iLSP provides security recommendations to their customers based on their extensive experience with various customers, for example route planning through dangerous zones, ideal travel times, or best practices for eliciting attention.

Importantly, iLSP offers 2 types of Monitoreo:

1. **Monitoreo Activo:** Monitoring is performed manually by a Monitorista, using external GPS devices and an external web page providing real-time location. No automatic alerts are generated, and all data is manually input into the system, making human error very common. A single Monitorista may be tracking from 10-30 vehicles at a time, and they must balance and prioritize their time between every trip. Any anomaly detected by a Monitorista should be verified with the vehicle Operator and manually captured as a ‘status’ in the iLSP system.
2. **Monitoreo Inteligente:** Monitoring is highly automated. iLSP-owned devices are installed in the transport vehicles, and automatically upload GPS data to iLSP systems. Alerts are generated by iLSP systems when anomalies occur, such as not following the pre-designated route, or stopped for longer than normal. When an alert is generated, the iLSP IVR system generates a phone call and assigns it to a Monitorista, who then contacts the vehicle Operator and enquires about the alert.

In both cases, any detected potential thefts are quickly transferred to the Mesa de Crisis team, who coordinate iLSP field teams and police to ensure the safety of the vehicle, or to recover the vehicle or goods if a theft has occurred.


## Section 2: Business Case Overview
iLSP wants to build an AI/ML powered solution that has the following business objectives:

1. Reduce the number of theft (or attempted theft) events that occur by taking preventative measures, thereby reducing the overall risk of theft for their customers.
2. Increase the efficiency of the Monitorista team, who are currently stretched thin and cannot pay full attention to all of their journeys in real-time (particularly Monitoreo Activo).
3. Generate more intelligent alerts and prioritize them, to reduce the number of unnecessary phone calls made to Operators by the Monitoristas (particularly Monitoreo Inteligente).

ILSP’s proposed AI/ML-powered solution could achieve this by:

1. Predicting probability of theft events in real-time, using live streamed GPS data. This will allow Monitoristas to pay closer attention to those journeys currently at risk, and to more regularly ensure that nothing has happened. Interpretability is important, as that will allow Monitoristas to provide valuable information to Customers and Operators about active risk factors.
2. Generate more intelligent alerts. Current alerts are rules-based, and every alert is checked with the Operator to ensure that nothing is going wrong. More intelligent alerts would potentially minimize the number of unnecessary calls, and maximize alerts when something is truly going wrong.
3. Prioritize alerts in real-time. Current alerts are generated and followed-through on a first-in-first-out basis, even if a more recent alert is much more critical. If alerts can be prioritized, this will ensure that critical alerts are followed-up more quickly.
4. **Building a predictive model for thefts using non-real-time data. This would allow iLSP to build a model that understands the inherent risk factors of journeys, so that they can plan journeys to minimize risk. E.G. traveling at a certain time of day, certain routes, identifying potentially colluded operators.**
5. Automating generation of Safety Reports using ML approach. Currently, manual reports are made, but iLSP would like a more powerful approach that could, for example, generate a report for customers once per week or per month about the safety summary of their trips (how safe were they, how many incidents, average response time of iLSP, how they could improve security, etc...).


**Additional Notes:**

* Some trips can be multiple days long, therefore theft might not be detected for a long time, especially is using manual monitoring systems (Monitoreo Activo). For example, 8 hours before a theft is detected.
* Long trips can mean planning routes can become complex. E.G. don’t want to travel through zone A at night, or zone B during the day.
* Risk of human loss of life during theft. Multiple deaths have happened, both to iLSP custodia staff, and Operators.
* Potentially, Operators or even employees of Customers may be colluding with the thieves to let them know which product is being carried, and what security measures are active. There is currently no analysis about who might be colluding based on frequency of thefts or products stolen.

```python
########################################################################################################
# THIS WAS PREVIOUS GEOHASHING ATTEMPT. WE SINCE CHANGED TO HILBERT-CURVE BASED GEOHASHING.
# Code is kept for reference.

# Now we are going to add geo-hashing to our dataframe, with 2, 3 and 4 digit precision.
# # This will allow us to easily group our journeys into discrete areas based on GPS coordinates.
# # This will help when post codes are not consistent, or are missing from the dataset.

# df_all['geohash_4'] = [geohash2.encode(coordinates[0], coordinates[1])[0:4] for coordinates in df_all[['latitude','longitude']].values]
# df_all['geohash_2'] = df_all['geohash_4'].str[0:2]
# df_all['geohash_3'] = df_all['geohash_4'].str[0:3]

# 4 digit geohashing with this library represents about 20km - 40km accuracy.
# Note that NaN geohashes are sent to 0000. 
```

```python
# In this cell, we are calculating a Hilbert-curve based Geohash to divide our country into segments (geohashes)
# This will allow us to represent 'proximity' to other thefts in our model, based on GPS, without relying on 
# post code which is sometimes unreliable.
# Precision of about 20km per geohash - https://github.com/tammoippen/geohash-hilbert

# This function takes the GPS coordinates (if available) and returns the geohash.
# If the GPS coordinates are missing, it uses the gmaps API to return 
# Thanks Oscar for writing the function (modified slightly by Jacob)
def get_geohash(longitude, latitude, municipality):
    # Try and get the geohash from the GPS coordinates
    try:
        geohash = ghh.encode(longitude,latitude, bits_per_char=2, precision=10)
        return geohash
    # If no GPS coords are provided, use gmaps API to get approximate GPS coords based on municipality
    # and use those coordinates to calculate geohash
    except:
        approx_gps_coords = gmaps.geocode(municipality)[0]["geometry"]["location"]
        geohash = ghh.encode(approx_gps_coords["lng"],approx_gps_coords["lat"], bits_per_char=2, precision=10)
        return geohash

# First we set the dataframe for the thefts (as we only have GPS for the theft events)
df_thefts = df_all[df_all['is_theft'] == 1]

# Get the geohash for each theft event and store in the df_thefts dataframe as a new column
df_thefts['geohash'] = [get_geohash(longitude=location[0], latitude=location[1],municipality=location[2]) \
               for location in df_thefts[['longitude', 'latitude','municipality']].values]

# Join the geohash column to the df_all dataframe of all trips
df_all = df_all.join(df_thefts['geohash'])
```

```python
# Drop the thefts where the GPS are NaN & the Municipality is 'nan', as this would return a random
# location using the 'nan' text. If we don't know where a theft occurred, we decide it is best to drop it.
df_all = df_all.drop(df_all[(df_all['latitude'].isna()) & (df_all['municipality'].isna()) &
                            (df_all['is_theft'] == 1)].index)
```

```python
# This function receives a df, a column, and corrects all instances of the name it is passed.
def fix_names(df, column, incorrect_name, correct_name):
    df = df.copy()
    for index in df[df[column] == incorrect_name].index:
        df.at[index, column] = correct_name
        print('changing {} in column {}: from "{}" to "{}"'.format(index, column, incorrect_name, correct_name))
    return df
```

# Building training and testing data sets

In order to build 

We will define our training data as all journeys starting before april 1st, 2019. Our testing data will consist of all data starting on 1st April or afterwards. Why are we splitting by time? Because we want this to simulate how we will really test this (data from the future).

How much should we split? That depends, a common rule of thumb is 70%/30% for training/testing. However, it also depends how much data you have. If you have a lot of data, you might do a 90%/10% split, as you want to use as much data as possible for training, and still have a representative testing set. 

In this case, we have decided on about 15 months of data for training, and 4 months for testing ~ 80%/20% train/test split. We have decided on this so that we still have some seasonality to the data, and capture the pattern of low january, high november etc... But this may not be the best split. In reality, we should not be touching the testing set at all once we have decided what it will be, in order not to bias our model for that testing set.

```python
# Some utilities to make our code cleaner

def calculate_dangerous_examples(df_train_thefts, column, minimum_counter):
    frequently_stolen_x = (df_train_thefts[column].value_counts() >= minimum_counter)
    frequently_stolen_x = frequently_stolen_x[frequently_stolen_x].index
    print('number of dangerous {}: {}'.format(str(column), len(frequently_stolen_x)))
    with open("./dangerous_examples/"+column+".txt", 'w') as f:
        for example in list(frequently_stolen_x):
            print(example, file=f)
    return list(frequently_stolen_x)

def one_hot_encode_column(df, column, dangerous_examples):
    ohe_all = pd.get_dummies(df[column])
    unwanted_columns = ohe_all.drop(columns=dangerous_examples).columns
    ohe = pd.get_dummies(df[column]).drop(columns=unwanted_columns)
    ohe = ohe.add_prefix(column+"_")
    df = pd.concat([df,ohe], axis=1, sort=False)
    return df
```

```python
# list of 'dangerous x' variables that we need to one-hot-encode
# customer_id
# product
# transport_line
# origin_id
# destination_id

df_all = one_hot_encode_column(df=df_all, column='customer_id', dangerous_examples=calculate_dangerous_examples(df_train_thefts, 'customer_id', 3))
df_all = one_hot_encode_column(df=df_all, column='product', dangerous_examples=calculate_dangerous_examples(df_train_thefts, 'product', 3))
df_all = one_hot_encode_column(df=df_all, column='transport_line', dangerous_examples=calculate_dangerous_examples(df_train_thefts, 'transport_line', 3))
df_all = one_hot_encode_column(df=df_all, column='origin_id', dangerous_examples=calculate_dangerous_examples(df_train_thefts, 'origin_id', 3))
df_all = one_hot_encode_column(df=df_all, column='destination_id', dangerous_examples=calculate_dangerous_examples(df_train_thefts, 'destination_id', 3))
```

```python
# unit type ID one hot encoding
df_all = pd.concat([df_all,pd.get_dummies(df_all['unit_type_id'])], axis=1, sort=False)

# start day of week one hot encoding
df_all = pd.concat([df_all,pd.get_dummies(df_all['start_dayofweek']).add_prefix('startdayofweek_')], axis=1,
                   sort=False)

# start month one hot encoding
df_all = pd.concat([df_all,pd.get_dummies(df_all['start_month']).add_prefix('month_')], axis=1, sort=False)

# monitoring_type one hot encoding
df_all = pd.concat([df_all,pd.get_dummies(df_all['monitoring_type_id'])], axis=1, sort=False).drop(columns=['activo'])

# Peak theft hours are from 4:00-10:59
peak_hours = [i for i in range(4, 11)]

# If a trip leaves during peak hours, set peak_start to 1
df_all['peak_start'] = df_all['start_time'].apply(lambda x: x.hour in peak_hours).astype(int)
```
